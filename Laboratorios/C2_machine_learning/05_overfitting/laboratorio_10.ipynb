## 1) Normalización datos

```python
df.dtypes #Verificamos que el tipo de datos con los que se trabaja
```

```python
x = df.drop(['id','diagnosis'],axis=1)
y = df.loc[:,['id','diagnosis']]

scaler=StandardScaler()
scaler.fit(x)
x_scaler=scaler.transform(x)

df_norm = pd.DataFrame(data=x_scaler, columns=x.columns)
df_norm
```

## 2) Gráfico de correlación

```python
fig = plt.figure(figsize = (25,25))

sns.heatmap(df_norm.corr().round(2),annot = True, fmt = 'g', cmap = 'gnuplot_r')             
plt.show()
```

```python
#Veamos mas claro que valores estan correlacionados con coeficiente mayor o igual a 0.9
fig = plt.figure(figsize = (25,25))
sns.heatmap(df_norm.corr().round(2),annot = True, fmt = 'g', cmap = 'gnuplot_r',vmin = 0.9,
                      vmax = 1,)             
plt.show()
```

## 3) PCA

```python
pca = PCA(n_components=10)

principalComponents = pca.fit_transform(df_norm.values) 

percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2)
columns = ['PC1', 'PC2', 'PC3', 'PC4',"PC5","PC6","PC7","PC8","PC9","PC10"]

plt.figure(figsize=(15,5))
plt.bar(x= range(1,11), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.show()
```

```python
percent_variance_cum = np.cumsum(percent_variance)
columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10']

plt.figure(figsize=(15,5))
plt.bar(x= np.arange(10), height=percent_variance_cum, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component Cumsum')
plt.title('PCA Scree Plot')
plt.show()
```

Del primer gráfico se puede notar que la primera componente es la que posee mayor varianza explicada (mayor a un $40\%$) y del segundo gráfico se puede notar que las primeras cuatro comonentes explican cerca del $80\%$ de la varianza mientras que las componentes $5, ,6 ,7 ,8, 9, 10$ explican solo el $20\%$ de la varianza.
